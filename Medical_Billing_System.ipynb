```python
import csv
import os

class Patient:
    def __init__(self, patient_id, name, insurance):
        self.patient_id = patient_id
        self.name = name
        self.insurance = insurance
        self.items = []

    def add_item(self, category, description, cost, covered=False):
        self.items.append({
            "category": category,
            "description": description,
            "cost": cost,
            "covered": covered
        })

    def bill(self):
        total = 0
        insurance_paid = 0
        patient_due = 0

        for item in self.items:
            total += item["cost"]

            if self.insurance and self.insurance.lower() != "no":
                insurance_paid = total * 0.8
                patient_due = total * 0.2
            else:
                insurance_paid = 0
                patient_due = total

        print("\n------ BILL ------")
        print("Patient:", self.name)
        print("Total Cost: $", total)
        print("Insurance Paid: $", insurance_paid)
        print("Patient Pays: $", patient_due)
        print("***---------Breakdown---------***")
        print("Consultation fee: $1500")
        print(f"charge for single day: $1000 and days stayed {days}")
        print("Room charge: $", 1000 * days)
        print("Medication cost: $", med_cost)

        file_exists = os.path.isfile("billing_records1.csv")

        with open("billing_records1.csv", "a", newline="") as f:
            writer = csv.writer(f)

            if not file_exists:
                writer.writerow([
                    "Patient ID",
                    "Name",
                    "Insurance",
                    "Total",
                    "Insurance Paid",
                    "Patient Due"
                ])

            writer.writerow([
                self.patient_id,
                self.name,
                self.insurance,
                total,
                insurance_paid,
                patient_due
            ])

        print("Record saved.")


# Main Program
patient_id = input("Patient ID / Registeration no: ")
name = input("Patient Name: ")
insurance = input("Insurance Provider: ")

p = Patient(patient_id, name, insurance)

# Consultation
consult_desc = input("Consultation Description: ")
p.add_item("Consultation", consult_desc, 1500)

# Room
room_desc = input("Room Description: ")
days = int(input("Number of Days: "))
p.add_item("Room", room_desc, 1000 * days)

# Medication
med_desc = input("Medication Description: ")
med_cost = float(input("Medication Cost: "))
covered = input("Insurance Covers Medication? (yes/no): ").lower() == "yes"
p.add_item("Medication", med_desc, med_cost, covered)

# Print bill and save
p.bill()
```
